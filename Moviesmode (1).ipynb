```python
movies=movies[['id', 'title', 'overview', 'genre']]
```

```python
movies
```

```python
movies['tags'] = movies['overview']+movies['genre'] # for new column 'tags'
```

```python
movies
```

```python
new_data  = movies.drop(columns=['overview', 'genre']) # to remove columns
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer # to extract features
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity # to implement cosine function
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"].index[0] # gives index of the movie
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand("Iron Man") # movies similar to iron man
```
